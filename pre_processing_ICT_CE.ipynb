```python
df['cat_lv1'] = df['tree_category'].str.split('->',expand=True)[0]
df['cat_lv2'] = df['tree_category'].str.split('->',expand=True)[1]
df['cat_lv3'] = df['tree_category'].str.split('->',expand=True)[2]
```

Lựa chọn bộ category lvl 3 có ít nhất 1000 sản phẩm mỗi category

```python
pd.set_option('display.max_rows', None)
dict_cat_lv3 = df['cat_lv3'].value_counts(ascending=True).to_dict()
final_cat_lv3 = []
for key, value in dict_cat_lv3.items():
    if value >= 25:
        final_cat_lv3.append(key)

print(final_cat_lv3)
```
